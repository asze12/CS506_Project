```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
state_census_purchase_df = pd.read_excel('State Census Purchase Combined.xlsx')
```

```python
state_census_purchase_df.head()
```

```python
state_census_purchase_df.fillna(0, inplace = True)
```

```python
political_aff = ['r' if val == 'Republican' else 'b' for val in state_census_purchase_df['Political Affiliation']]
```

```python
state_census_purchase_df
```

# Population Density

```python
fig, ax = plt.subplots(3, 1, figsize=(15,24))
x_data = state_census_purchase_df['Population']
y_weapons_data = np.array(state_census_purchase_df['Weapons Purchase Amount'])/1000
y_accessory_data = np.array(state_census_purchase_df['Tactical Equipment Purchase Amount'])/1000
y_vehicle_data = np.array(state_census_purchase_df['Vehicle Purchase Amount'])/1000

ax[0].scatter(x_data, y_weapons_data, c = political_aff, label = 'Weapons Total Purchase')
#ax[0].set_xscale('log')
ax[0].set_title('Weapon Acquisition Value vs. Population Density', fontsize = 15)
ax[0].set_xlabel('State Population Density', fontsize = 15)
ax[0].set_ylabel('Weapons Acquisition Value Amount (x $1000)', fontsize = 15)
ax[0].grid()
#ax[0].legend()

ax[1].scatter(x_data, y_accessory_data, c = political_aff, label = 'Tactical Equipment Total Purchase')
#ax[1].set_xscale('log')
ax[1].set_title('Tactical Equipment Acquisition Value vs. Population Density', fontsize = 15)
ax[1].set_xlabel('State Population Density', fontsize = 15)
ax[1].set_ylabel('Tactical Equipment Acquisition Amount (x $1000)', fontsize = 15)
ax[1].grid()
#ax[1].legend()

ax[2].scatter(x_data, y_vehicle_data, c = political_aff, label = 'Vehicle Total Purchase')
#ax[2].set_xscale('log')
ax[2].set_title('Vehicle Acquisition Value vs. Population Density', fontsize = 15)
ax[2].set_xlabel('State Population Density', fontsize = 15)
ax[2].set_ylabel('Vehicle Acquisition Amount (x $1000)', fontsize = 15)
ax[2].grid()
#ax[2].legend()

# ax2 = ax.twinx()
# ax2.scatter(x_data, y_vehicle_data, c = 'g', label = 'Vehicle Total Purchase')
# ax2.set_ylabel(' Vehicle Purchase Amount (x $1000)')

plt.show()
```
