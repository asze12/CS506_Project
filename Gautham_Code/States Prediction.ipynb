```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn import neural_network
```

```python
states_data = pd.read_excel('State Census Purchase Combined.xlsx')
states_data
```

```python
# x_columns = ['Population Density', 'HS Completion Rate (%)', 'Higher Education (Bachlors or Higher) (%)', 'Median Household Income ($)',
#             'Poverty Rate (%)', 'Black (%)', 'America Indian (%)', 'Asian (%)', 'Hispanic (%)', 'White (%)']


x_columns = ['Black (%)',  '# of Violent Crimes (2016)']

xdata = states_data[x_columns]
ydata = states_data['Weapons Purchase Amount']

linmodel = linear_model.LinearRegression(fit_intercept = True, normalize = False)
```

```python
#xdata = np.array(xdata)
#ydata = np.array(ydata)
#totaldata = np.hstack((xdata, ydata))
#np.random.shuffle(xdata)
training_xdata = np.array(xdata)[:30, :]
training_ydata = np.array(ydata)[:30]
testing_xdata = np.array(xdata)[30:, :]
testing_ydata = np.array(ydata)[30:]
```

```python
linmodel.fit(training_xdata, training_ydata)
linmodel.score(testing_xdata, testing_ydata)
```
