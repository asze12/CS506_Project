# Importing Mass Data with Department, Census Data Name

```python
mass_data = pd.read_excel('Mass_Data.xlsx')
mass_data.dropna(axis=0, how='any', inplace = True)
mass_data.reset_index(drop=True, inplace=True)
mass_pop_hou = pd.read_csv('./MASS_data/Mass_cities_pop_housing.csv', header = 1)
mass_pop_hou = mass_pop_hou[['Geographic area.1', 'Density per square mile of land area - Population', 'Density per square mile of land area - Housing units']]
```

```python
mass_data.head()
```

```python
mass_city_hs = pd.read_csv('./MASS_data/mass_city_HighSchool.csv', header = 1)
mass_pov = pd.read_csv('./MASS_data/mass_poverty.csv', header = 1)
mass_race_br = pd.read_csv('./MASS_data/mass_race_breakdown.csv', header = 1)
mass_income = pd.read_csv('./MASS_data/mass_city_median_income.csv', header = 1)
```

```python
race_list = ['Percent of total population - Race - One Race - White', 'Percent of total population - Race - One Race - Black or African American', 'Percent of total population - Race - One Race - American Indian and Alaska Native', 'Percent of total population - Race - One Race - Asian', 'Percent of total population - Race - One Race - Native Hawaiian and Other Pacific Islander', 'Percent of total population - Race - One Race - Some Other Race', 'Percent of total population - Race - Two or More Races']
```

```python
mass_income
```

```python
mass_pop_den = []
mass_hou_den = []
mass_edu_hs = []
mass_pov_p = []
mass_income_hhold =[]
mass_income_per_cap = []
mass_race = pd.DataFrame()
for town in mass_data['Census Name']:
    mass_pop_den.append(mass_pop_hou[mass_pop_hou['Geographic area.1'] == town]['Density per square mile of land area - Population'].tolist()[0])
    
    mass_hou_den.append(mass_pop_hou[mass_pop_hou['Geographic area.1'] == town]['Density per square mile of land area - Housing units'].tolist()[0])
    
    mass_edu_hs.append(mass_city_hs[mass_city_hs['Geographic Area1'] == town]['Percent'].tolist()[0])
    
    mass_pov_p.append(mass_pov[mass_pov['Geographic Area1'] == town]['Percent'].tolist()[0])
    
    mass_income_hhold.append(mass_income[mass_income['Geographical Area1'] == town]['Median income in 1999 (dollars) - Households'].tolist()[0])
    mass_income_per_cap.append(mass_income[mass_income['Geographical Area1'] == town]['Per capita income in 1999 (dollars)'].tolist()[0])
    
    mass_race = mass_race.append(mass_race_br[mass_race_br['Geographic Area1'] == town][race_list], ignore_index = True)
    
mass_data['Density per sq. mile - Pop'] = mass_pop_den
mass_data['Density per sq. mile - Housing'] = mass_hou_den
mass_data['HS Completion Percent'] = mass_edu_hs
mass_data['Poverty Percent'] = mass_pov_p
mass_data['Median income in 1999 (dollars) - Households'] = mass_income_hhold
mass_data['Per capita income in 1999 (dollars)'] = mass_income_per_cap
mass_data = mass_data.join(mass_race, how = 'left')
```

```python
mass_data_lm = pd.read_pickle('mass_data.pkl')
```

```python
dept_total_items = []
dept_avg_cost = []
dept_total_cost = []
dept_median_cost = []
for dept in dept_acq_d:
    dept_name = dept['Station Name (LEA)'].unique()
    if dept_name not in ['BERKSHIRE COUNTY SHERIFFS OFFICE', 'BRISTOL COUNTY SHERIFF DEPT','CENTRAL MA LAW ENFORCEMENT']:
        dept_total_items.append(dept['Quantity'].sum())
        dept_avg_cost.append(((dept['Acquisition Value']*dept['Quantity']).sum())/dept['Quantity'].sum())
        dept_total_cost.append((dept['Acquisition Value']*dept['Quantity']).sum())
        dept_median_cost.append(dept['Acquisition Value'].median())
```

```python
mass_data_lm['Total Items Recieved'] = dept_total_items
mass_data_lm['Average Cost per Item'] = dept_avg_cost
mass_data_lm['Median Cost'] = dept_median_cost
mass_data_lm['Total Cost'] = dept_total_cost
```

```python
dept_dict = []
for dept in dept_acq_d:
    dept_name = dept['Station Name (LEA)'].unique()
    if dept_name not in ['BERKSHIRE COUNTY SHERIFFS OFFICE', 'BRISTOL COUNTY SHERIFF DEPT','CENTRAL MA LAW ENFORCEMENT']:
        dept_weapon_acq = dept.groupby('Item Name')
        item_l = []
        quan_l = []
        ind_dept = {}
        for item, quan in dept_weapon_acq:
            if item in imp_items:
                ind_dept.update({item:sum(quan['Quantity'].tolist())})
        dept_dict.append(ind_dept)
```

```python
from sklearn.feature_extraction import DictVectorizer

v = DictVectorizer(sparse = False)
X = v.fit_transform(dept_dict)
```

```python
for i, item in enumerate(v.get_feature_names()):
    mass_data[item] = list(X[:,i])
```
